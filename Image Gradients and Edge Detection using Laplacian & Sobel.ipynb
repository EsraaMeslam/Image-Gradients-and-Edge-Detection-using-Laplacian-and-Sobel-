```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python

img = cv2.imread("messi.webp", 0)# Read the image

lap = cv2.Laplacian(img, cv2.CV_64F, ksize=3)# Laplacian filter to detect edges
lap = np.uint8(np.absolute(lap))


sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)# Sobel filter in the X direction to detect horizontal edges
sobelX = np.uint8(np.absolute(sobelX))


sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)# Sobel filter in the Y direction to detect vertical edges
sobelY = np.uint8(np.absolute(sobelY))


sobelCombined = cv2.bitwise_or(sobelX, sobelY)#OR between sobel x , sobel y

titles = ['Original image', 'Laplacian', 'Sobel_X', 'Sobel_Y', 'Sobel Combined']


images = [img, lap, sobelX, sobelY, sobelCombined]


plt.figure(figsize=(20, 10))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])

plt.show()
```

Convert the result to unsigned 8-bit integer
This is necessary because cv2.Laplacian(),Sobel_X,Sobel_Y may produce negative values
and converting them to uint8 allows us to visualize them correctly

```python

img = cv2.imread("sudoku.png", 0)# Read the image




sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)# Sobel filter in the X direction to detect horizontal edges
sobelX = np.uint8(np.absolute(sobelX))


sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)# Sobel filter in the Y direction to detect vertical edges
sobelY = np.uint8(np.absolute(sobelY))


sobelCombined = cv2.bitwise_or(sobelX, sobelY)#OR between sobel x , sobel y


titles = ['Original image', 'Sobel_X\n(Vertical edges)', 'Sobel_Y\n(Horizontal edges)', 'Sobel Combined\n(Vertical&Horizontal edges)']

images = [img, sobelX, sobelY, sobelCombined]


plt.figure(figsize=(20, 10))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])

plt.show()
```
